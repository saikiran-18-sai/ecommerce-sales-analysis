```python


import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/ecommerce_sales.csv")

#info
df.info()

df.describe()
#shape of the data set

df.shape



```

```python
df.columns
df.isnull().sum()
df['Category'].unique()
```

```python
df.duplicated().sum()
''' if duplicates exist 
df.drop_duplicates'''
df.isnull().sum()
```

```python
df['Calculated_Sale'] = df['Quantity'] * df['Price_p' \
'er_Unit']
df[['Total_Sale','Calculated_Sale']].head()
```

```python

category_sales = df.groupby('Category')['Total_Sale'].sum()

category_sales.plot(kind='bar')

plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")

plt.show()
```

```python
top_products = df.groupby('Product')['Total_Sale'].sum().sort_values(ascending=False).head(5)

top_products.plot(kind='bar')

plt.title("Top 5 Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Total Sales")

plt.show()
```

```python
sns.histplot(df['Quantity'])

plt.title("Distribution of Quantity Sold")

plt.show()
```

```python
sns.histplot(df['Price_per_Unit'])

plt.title("Price Distribution")

plt.show()
```
